```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
#1.Fill in these missing numbers and make the column an integer column(instead of a float column).
for i in df['FlightNumber'].index:
        if pd.isnull(df['FlightNumber'][i]) :
            val=df['FlightNumber'][i-1]
            a=val+10
            df.loc[i,'FlightNumber']=a
            
df = df.astype({"FlightNumber": int})
```

```python
df_temp=df.copy(deep=True)
```

```python
df_temp
```

```python
#The From_To column would be better as two separate columns! Split each string on the underscore delimiter _ to give a new temporary DataFrame with
lst=df['From_To']
j=0
for i in lst:
        new=i.split("_")
        df_temp.loc[j,'From Location']=new[0]
        df_temp.loc[j,'To Location']=new[1]
        j=j+1  
df_temp['To Location'] = df_temp['To Location'].str.capitalize() 
df_temp['From Location'] = df_temp['From Location'].str.capitalize() 
```

```python
df_temp
```

```python
#Expand the Series of lists into a DataFrame named delays, rename the columnsdelay_1, delay_2, etc. and replace the unwanted RecentDelays column in dfwith delays.
df=df_temp
del df['From_To']
delay=[]
for i in df['RecentDelays'].index:
    delay=df['RecentDelays'][i]
    a=len(delay)
    for j in range(a):
        df.loc[i,'RecentDelays'+ str(j)]=delay[j]
```
