```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import scipy.misc
```

```python
f = scipy.misc.face(gray=True)
```

```python
f
```

```python
rows,cols = f.shape
```

```python
fr=f.reshape(rows*cols,1)
fr.shape
```

```python
plt.figure(figsize=(10, 3.6))
plt.subplot(131)
plt.imshow(f, cmap=plt.cm.gray)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y= kmeans.fit(fr)
y.labels_
```

```python
compressed_image = kmeans.cluster_centers_[y.labels_]
img=compressed_image.reshape(rows,cols)
```

```python
img
```

```python
plt.imshow(img, cmap=plt.cm.gray)
```
